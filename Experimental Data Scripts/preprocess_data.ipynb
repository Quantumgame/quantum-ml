```python
# This notebook is intended to display experimental data and preprocess it to be fed to the neural network

# This box is used to create png files of the data and store it in a folder
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
%matplotlib

data_folder_path = "/Users/sandesh/exp_data/"
map_folder_path = "/Users/sandesh/exp_data/maps/"
files = glob.glob(data_folder_path + "*.dat")

# Data format is V_LGD I_DC(nA) V_LGS I_AC(nA) t(sec)
# The format of the loaded array is [num_points,5]
for file in files:
    plt.clf()
    dat = np.loadtxt(file)
    plt.tricontourf(dat[:,0],dat[:,2],dat[:,1])
    plt.xlabel(r'$V_{LGD}$ (V)',fontsize=16)
    plt.ylabel(r'$V_{LGS}$ (V)',fontsize=16)
    cbar = plt.colorbar()
    cbar.set_label(r'$I_{DC}$ (nA)',fontsize=16)
    # -3 is a simple way to remove the dat extension
    # . is already present in the filename, so only png needs to be appended
    plt.savefig(os.path.join(map_folder_path,os.path.basename(file)[:-3] + "png"))
```
