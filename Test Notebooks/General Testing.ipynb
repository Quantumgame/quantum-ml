# Testing of Sampling from a Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
def random_sample(mean,sigma_mu=0.05):
    return np.random.normal(mean,sigma_mu*mean)

mean = 1.0
print(random_sample(mean))
```

```python
N = 10000
samples = np.array([random_sample(mean) for x in range(N)])
freq,val = np.histogram(samples,bins="auto")
plt.plot(val[:-1],freq/N)
plt.xlabel('x')
plt.ylabel('Freq. (x)')
plt.title('Histogram for sampling from a Gaussian')
```
