```python
#This notebook will be used to test features for logging and TensorBoard in TensorFlow

import os

import numpy as np
import tensorflow as tf

tf.logging.set_verbosity(tf.logging.INFO)

IRIS_TRAINING = os.path.join(os.path.dirname("__file__"), "iris_training.csv")
IRIS_TEST = os.path.join(os.path.dirname("__file__"),"iris_test.csv")

def main(unused_argv):
    training_set = tf.contrib.learn.datasets.base.load_csv_with_header(
        filename=IRIS_TRAINING, target_dtype=np.int, features_dtype=np.float32)
    test_set = tf.contrib.learn.datasets.base.load_csv_with_header(
        filename=IRIS_TEST, target_dtype=np.int, features_dtype=np.float32)
    
    feature_columns = [tf.contrib.layers.real_valued_column("",dimension=4)]
    
    classifier = tf.contrib.learn.DNNClassifier(feature_columns=feature_columns,
                                               hidden_units=[10,20,10],
                                               n_classes=3,
                                               model_dir="/tmp/iris_model/",
                                               config=tf.contrib.learn.RunConfig(save_checkpoints_secs=1))
    
    def get_train_input():
        x = tf.constant(training_set.data)
        y = tf.constant(training_set.target)
        return x,y
    def get_test_input():
        x = tf.constant(test_set.data)
        y = tf.constant(test_set.target)
        return x,y 
    
    validation_monitor = tf.contrib.learn.monitors.ValidationMonitor(
        test_set.data,
        test_set.target,
        every_n_steps=100
        )
    classifier.fit(input_fn=get_train_input,
                   steps=2000,
                   monitors=[validation_monitor]
                  )
    
    
  
    accuracy_score= classifier.evaluate(input_fn=get_test_input,steps=1)["accuracy"]
    
 
    
    print("Accuracy",accuracy_score)
    
    new_samples = np.array(
            [[6.4,3.2,4.5,1.5],[5.0,3.1,5.1,1.7]],dtype=np.float32)
    
    y = classifier.predict(new_samples, as_iterable=False)
    print("Predictions",y)
    
if __name__ == "__main__":
    tf.app.run()
```
